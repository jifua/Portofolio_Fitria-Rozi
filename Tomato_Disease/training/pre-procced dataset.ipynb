```python
import os
import random
import shutil

# Tentukan path dataset di Google Drive
dataset_path = '/content/drive/MyDrive/Preprocessed data'  # Lokasi dataset asli
output_folder = '/content/drive/MyDrive/DATASET TOMAT '  # Folder output untuk dataset yang sudah dibagi

# Tentukan persentase data untuk train, validasi, dan test
train_ratio = 0.7
val_ratio = 0.2
test_ratio = 0.1

# Membuat folder train, val, dan test di output_folder
os.makedirs(os.path.join(output_folder, 'train'), exist_ok=True)
os.makedirs(os.path.join(output_folder, 'val'), exist_ok=True)
os.makedirs(os.path.join(output_folder, 'test'), exist_ok=True)

# Mendapatkan semua folder kategori (misalnya kelas penyakit daun) di dalam dataset
categories = os.listdir(dataset_path)

# Membuat subfolder untuk setiap kategori di dalam folder train, val, dan test
for category in categories:
    os.makedirs(os.path.join(output_folder, 'train', category), exist_ok=True)
    os.makedirs(os.path.join(output_folder, 'val', category), exist_ok=True)
    os.makedirs(os.path.join(output_folder, 'test', category), exist_ok=True)

    # Dapatkan semua file gambar dalam kategori ini
    category_path = os.path.join(dataset_path, category)
    images = os.listdir(category_path)

    # Acak urutan gambar
    random.shuffle(images)

    # Tentukan jumlah gambar untuk train, val, dan test
    train_size = int(len(images) * train_ratio)
    val_size = int(len(images) * val_ratio)

    # Bagi gambar ke train, val, dan test
    train_images = images[:train_size]
    val_images = images[train_size:train_size + val_size]
    test_images = images[train_size + val_size:]

    # Pindahkan gambar ke folder yang sesuai
    for image in train_images:
        shutil.copy(os.path.join(category_path, image), os.path.join(output_folder, 'train', category, image))
    for image in val_images:
        shutil.copy(os.path.join(category_path, image), os.path.join(output_folder, 'val', category, image))
    for image in test_images:
        shutil.copy(os.path.join(category_path, image), os.path.join(output_folder, 'test', category, image))

print("Pembagian dataset selesai!")
```

```python
import os

# Fungsi untuk menghitung jumlah file dalam suatu folder
def count_files_in_folder(folder_path):
    total_files = 0
    for root, dirs, files in os.walk(folder_path):
        total_files += len(files)
    return total_files

# Path ke folder train, val, dan test
train_folder = '/content/drive/MyDrive/DATASET TOMAT /train'
val_folder = '/content/drive/MyDrive/DATASET TOMAT /val'
test_folder = '/content/drive/MyDrive/DATASET TOMAT /test'

# Menghitung jumlah file dalam setiap folder
train_count = count_files_in_folder(train_folder)
val_count = count_files_in_folder(val_folder)
test_count = count_files_in_folder(test_folder)

# Menampilkan hasil
print(f"Jumlah gambar di folder TRAIN: {train_count}")
print(f"Jumlah gambar di folder VALIDATION: {val_count}")
print(f"Jumlah gambar di folder TEST: {test_count}")
```

```python
import os

# Fungsi untuk menghitung jumlah gambar dalam setiap kelas (subfolder)
def count_images_per_class(folder_path):
    class_counts = {}
    for class_name in os.listdir(folder_path):
        class_folder = os.path.join(folder_path, class_name)
        if os.path.isdir(class_folder):  # Pastikan ini folder, bukan file
            class_counts[class_name] = len(os.listdir(class_folder))
    return class_counts

# Path ke folder train, val, dan test
train_folder = '/content/drive/MyDrive/DATASET TOMAT /train'
val_folder = '/content/drive/MyDrive/DATASET TOMAT /val'
test_folder = '/content/drive/MyDrive/DATASET TOMAT /test'

# Menghitung jumlah gambar per kelas di setiap set
train_class_count = count_images_per_class(train_folder)
val_class_count = count_images_per_class(val_folder)
test_class_count = count_images_per_class(test_folder)

# Menampilkan hasil
print("Jumlah gambar per kelas di folder TRAIN:")
for class_name, count in train_class_count.items():
    print(f"{class_name}: {count} gambar")

print("\nJumlah gambar per kelas di folder VALIDATION:")
for class_name, count in val_class_count.items():
    print(f"{class_name}: {count} gambar")

print("\nJumlah gambar per kelas di folder TEST:")
for class_name, count in test_class_count.items():
    print(f"{class_name}: {count} gambar")
```
