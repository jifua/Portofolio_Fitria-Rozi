```python
# Import libraries yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.utils import plot_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator  # Untuk augmentasi gambar
from tensorflow.keras.applications import MobileNet  # Ganti dengan MobileNet
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization  # Layer untuk model kustom
from tensorflow.keras.models import Model  # Model untuk menggabungkan layer
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau  # Callback untuk menghentikan pelatihan dan menyesuaikan learning rate
from tensorflow.keras.optimizers import SGD  # Optimizer SGD
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, f1_score  # Untuk evaluasi model
import seaborn as sns  # Untuk visualisasi confusion matrix
from tensorflow.keras.regularizers import l2  # Regularizer untuk mencegah overfitting
from sklearn.preprocessing import label_binarize  # Untuk binarisasi label pada ROC curve

# Path ke direktori dataset
train_dir = '/content/drive/MyDrive/Tomato Leaf Disease/train'  # Path untuk data train
val_dir = '/content/drive/MyDrive/Tomato Leaf Disease/val'  # Path untuk data validasi
test_dir = '/content/drive/MyDrive/Tomato Leaf Disease/test'  # Path untuk data tes

# Data augmentation untuk meningkatkan variasi data pelatihan
train_datagen = ImageDataGenerator(
    rescale=1./255,  # Menormalisasi gambar
    shear_range=0.2,  # Menyebabkan pemotongan (shear) gambar secara acak
    zoom_range=0.2,  # Zoom gambar secara acak
    rotation_range=45,  # Rotasi gambar secara acak
    width_shift_range=0.2,  # Geser gambar secara horizontal
    height_shift_range=0.2,  # Geser gambar secara vertikal
    horizontal_flip=True,  # Pembalikan gambar secara horizontal
    vertical_flip=True,  # Pembalikan gambar secara vertikal
    brightness_range=[0.7, 1.3],  # Variasi kecerahan gambar
    fill_mode='nearest'  # Mengisi piksel kosong dengan piksel terdekat
)

# Untuk data validasi dan tes, hanya dilakukan normalisasi gambar
val_datagen = ImageDataGenerator(rescale=1./255)  # Normalisasi untuk validasi
test_datagen = ImageDataGenerator(rescale=1./255)  # Normalisasi untuk tes

# Menggunakan generator untuk membaca gambar dari direktori
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),  # Ukuran input untuk MobileNet
    batch_size=16,  # Ukuran batch
    class_mode='categorical'  # Karena klasifikasi multikelas
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(224, 224),  # Ukuran input untuk MobileNet
    batch_size=16,  # Ukuran batch
    class_mode='categorical'  # Mode kategorikal
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),  # Ukuran input untuk MobileNet
    batch_size=16,  # Ukuran batch
    class_mode='categorical',  # Untuk evaluasi, menggunakan kategori
    shuffle=False  # Jangan acak data tes untuk evaluasi yang tepat
)

# Memuat model MobileNet dengan classifier (top layer) dan menggunakan bobot yang sudah dilatih di ImageNet
mobilenet = MobileNet(
    include_top=True,  # Menyertakan lapisan classifier
    weights='imagenet',  # Menggunakan bobot yang sudah dilatih di ImageNet
    input_shape=(224, 224, 3),  # Ukuran input untuk MobileNet
    pooling=None,  # Tidak ada pooling pada akhir model
    classes=1000,  # Jumlah kelas pada ImageNet
    classifier_activation="softmax"  # Menggunakan softmax untuk klasifikasi
)

# Membekukan beberapa lapisan dari model MobileNet agar tidak diubah selama pelatihan
for layer in mobilenet.layers[:10]:  # Membekukan 10 lapisan pertama
    layer.trainable = False

# Menambahkan lapisan kustom untuk klasifikasi di atas model MobileNet
x = Flatten(name='flatten_custom')(mobilenet.output)  # Meratakan output dari MobileNet
x = Dense(512, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)  # Lapisan Dense dengan regularisasi L2
x = BatchNormalization()(x)  # Normalisasi batch untuk stabilitas pelatihan
x = Dropout(0.3, name='dropout_1')(x)  # Dropout untuk mengurangi overfitting
x = Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)  # Lapisan Dense kedua
x = BatchNormalization()(x)  # Normalisasi batch lagi
x = Dropout(0.5, name='dropout_2')(x)  # Dropout lagi untuk mengurangi overfitting
output = Dense(len(train_generator.class_indices), activation='softmax')(x)  # Lapisan output untuk klasifikasi multikelas

# Menyusun model dengan input dari MobileNet dan output dari lapisan kustom
model_mobilenet = Model(inputs=mobilenet.input, outputs=output)

# Menampilkan arsitektur model
model_mobilenet.summary()

# Menyimpan arsitektur model sebagai file PNG
plot_model(model_mobilenet, to_file='model_mobilenet_architecture.png', show_shapes=True, show_layer_names=True)
print("Model summary telah disimpan sebagai 'model_mobilenet_architecture.png'.")


# Mengkompilasi model dengan optimizer SGD dan menggunakan loss categorical crossentropy
model_mobilenet.compile(optimizer=SGD(learning_rate=1e-4, momentum=0.9),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Menyusun callback untuk pelatihan
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)  # Menghentikan pelatihan jika validasi loss tidak membaik
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6, verbose=1)  # Mengurangi learning rate jika validasi loss stagnan

# Melatih model dengan data pelatihan dan validasi
history_mobilenet = model_mobilenet.fit(
    train_generator,
    validation_data=val_generator,
    epochs=50,  # Jumlah epoch pelatihan
    callbacks=[early_stopping, reduce_lr]  # Callback yang digunakan
)

# Mengevaluasi model dengan data tes
test_loss_mobilenet, test_acc_mobilenet = model_mobilenet.evaluate(test_generator)  # Evaluasi model pada data tes
print(f'Test Accuracy: {test_acc_mobilenet * 100:.2f}%')  # Menampilkan akurasi tes
print(f'Test Loss: {test_loss_mobilenet :.4f}%')  # Menampilkan loss tes

# Memvisualisasikan sejarah pelatihan (akurasi dan loss)
plt.figure(figsize=(14, 5))

# Plot akurasi
plt.subplot(1, 2, 1)
plt.plot(history_mobilenet.history['accuracy'], label='Train Accuracy')  # Akurasi pelatihan
plt.plot(history_mobilenet.history['val_accuracy'], label='Validation Accuracy')  # Akurasi validasi
plt.title('Model Accuracy')  # Judul plot
plt.xlabel('Epoch')  # Label sumbu X
plt.ylabel('Accuracy')  # Label sumbu Y
plt.legend()  # Menampilkan legenda

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history_mobilenet.history['loss'], label='Train Loss')  # Loss pelatihan
plt.plot(history_mobilenet.history['val_loss'], label='Validation Loss')  # Loss validasi
plt.title('Model Loss')  # Judul plot
plt.xlabel('Epoch')  # Label sumbu X
plt.ylabel('Loss')  # Label sumbu Y
plt.legend()  # Menampilkan legenda

plt.show()

# Matriks kebingungannya
y_true = test_generator.classes  # Nilai yang benar dari data tes
y_pred = np.argmax(model_mobilenet.predict(test_generator), axis=-1)  # Prediksi kelas yang memiliki probabilitas tertinggi

cm = confusion_matrix(y_true, y_pred)  # Menghitung confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=test_generator.class_indices.keys(), yticklabels=test_generator.class_indices.keys())  # Visualisasi confusion matrix
plt.xlabel('Predicted')  # Label sumbu X
plt.ylabel('True')  # Label sumbu Y
plt.title('Confusion Matrix')  # Judul plot
plt.show()

# Menampilkan classification report
print('Classification Report:')
print(classification_report(y_true, y_pred, target_names=test_generator.class_indices.keys()))  # Laporan klasifikasi

# Menghitung F1 score
f1 = f1_score(y_true, y_pred, average='weighted')  # Menghitung F1 score rata-rata tertimbang
print(f'F1 Score: {f1:.4f}')  # Menampilkan F1 score

# Menghitung ROC Curve untuk setiap kelas
y_test_bin = label_binarize(y_true, classes=list(range(len(test_generator.class_indices))))  # Binarisasi label untuk ROC curve
y_pred_prob = model_mobilenet.predict(test_generator)  # Probabilitas prediksi

# Plot ROC curve untuk setiap kelas
plt.figure(figsize=(14, 10))
for i in range(len(test_generator.class_indices)):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_pred_prob[:, i])  # Menghitung false positive rate dan true positive rate
    roc_auc = auc(fpr, tpr)  # Menghitung area under the curve (AUC)
    plt.plot(fpr, tpr, label=f'Class {list(test_generator.class_indices.keys())[i]} (AUC = {roc_auc:.2f})')  # Mem-plot setiap kelas

plt.plot([0, 1], [0, 1], 'k--')  # Plot garis diagonal (AUC = 0.5)
plt.title('ROC Curve for Each Class')  # Judul plot ROC curve
plt.xlabel('False Positive Rate')  # Label sumbu X
plt.ylabel('True Positive Rate')  # Label sumbu Y
plt.legend(loc='best')  # Menampilkan legenda
plt.show()

# Menyimpan model dalam format HDF5
model_mobilenet.save('/content/drive/MyDrive/saved_model/mobilenetpengujian2_tomato_disease.h5')  # Menyimpan model
```
